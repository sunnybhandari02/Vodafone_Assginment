# Task 1 :- Identifying and importing essential libraries

```python
# Importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Task 3 :- Find out the per unit price from the data

```python
# Convert date time field into pandas time object
df['Order Date'] = df['Order Date'].dt.strftime('%Y/%m/%d')
df['Ship Date'] = df['Ship Date'].dt.strftime('%Y/%m/%d')
```

```python
df.dtypes
```

```python
# creating a columns of price per unit from the data
df['Price Per Unit'] = df['Sales'] / df['Quantity']
```

```python
df.head()
```

# Task 4 :- Find out the monthly revenue and analyze the findings

```python
# Creating a column with Year and month field only
df['year_month'] = pd.to_datetime(df['Order Date']).dt.to_period('M')
df.head()
```

```python
# Plotting a bar graph for checking monthly sales
plt.figure(figsize=(50,12))
t = df.groupby(['year_month'])['Sales'].sum()
ax = sns.barplot(data = pd.DataFrame(t).transpose())
ax.set(xlabel= "year_month", ylabel="Sales")
```

From the above plot we found that Sales were maximum on "2017-11" and minimum on "2014-02".

# Task 5 :- Find out the yearly revenue and analyze the findings

```python
# creating a column with year field
df['year'] = pd.to_datetime(df['Order Date']).dt.to_period('Y')
df.head()
```

```python
df.groupby('year')['Sales'].agg(['sum']).plot.bar()
```

From the above bar plot, we found that sales of the store has increased ever year.

# Task 6 :- Finding out the monthly growth rate and analyse the findings

```python
# Creating a column for month.
df['month'] = pd.DatetimeIndex(df['Order Date']).month
```

```python
# Checking monthly growth rate.
df.groupby('month')['Sales'].agg(['sum']).plot.bar()
```

From the above bar plot, we found that sales were very high in the last months of the year.

# Task 11 :- Find out the most and least sold product category from the store

```python
# a.) Value based
t = df.groupby(['Category'])['Sales'].sum()
ax = sns.barplot(data = pd.DataFrame(t).transpose())
ax.set(xlabel= "Category", ylabel="Sales")
```

Therefore, most sold product category is "Technology" and least sold product category is "Office Supplies Category".

```python
# b.) Quantity based
t = df.groupby(['Category'])['Quantity'].sum()
ax = sns.barplot(data = pd.DataFrame(t).transpose())
ax.set(xlabel= "Category", ylabel="Quantity")
```

Therefore, most sold product category on the basis of quantity is "Office Supplies Category" and least sold category is "Technology".

# Task 12 :- Find out the most and least sold product sub category from the
# store

```python
# a.) Value Based
plt.figure(figsize=(20,12))
t = df.groupby(['Sub-Category'])['Sales'].sum()
ax = sns.barplot(data = pd.DataFrame(t).transpose())
ax.set(xlabel= "Sub-Category", ylabel="Sales")
```

Therefore, most sold product sub category are "Chairs" and "Phones" and least sold product sub category is "Fastners".

```python
# b.) Quantity Based
plt.figure(figsize=(20,12))
t = df.groupby(['Sub-Category'])['Quantity'].sum()
ax = sns.barplot(data = pd.DataFrame(t).transpose())
ax.set(xlabel= "Sub-Category", ylabel="Quantity")
```

Therefore, most sold product sub category is "Binders" and least sold product sub category is "Copiers".
